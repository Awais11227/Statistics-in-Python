```python
import numpy as np

# Generate samples from two normal distributions
sample1 = np.random.normal(loc=20, scale=5, size=300)
sample2 = np.random.normal(loc=40, scale=5, size=700)

# Concatenate both samples into a single dataset
sample = np.hstack((sample1, sample2))

# Optional: Display a histogram to visualize the distribution
import matplotlib.pyplot as plt

plt.hist(sample, bins=30, edgecolor='black', alpha=0.7)
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram of Combined Sample")
plt.show()
```

```python
# plot histogram bins=50
plt.hist(sample,bins=50)
```

```python
from sklearn.neighbors import KernelDensity

model = KernelDensity(bandwidth=5, kernel='gaussian')

# convert data to a 2D array
sample = sample.reshape((len(sample), 1))

model.fit(sample)
```

```python
values = np.linspace(sample.min(),sample.max(),100)
values = values.reshape((len(values), 1))
```

```python
probabilities = model.score_samples(values)
probabilities = np.exp(probabilities)
```

```python
plt.hist(sample, bins=50, density=True)
plt.plot(values[:], probabilities)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Ensure 'species' is categorical
df['species'] = df['species'].astype(str)

# KDE Plot (Density Plot)
sns.kdeplot(data=df, x='petal_width', hue='species', fill=True)

# ECDF Plot (Empirical Cumulative Distribution Function)
sns.ecdfplot(data=df, x='petal_width', hue='species')

plt.show()
```
